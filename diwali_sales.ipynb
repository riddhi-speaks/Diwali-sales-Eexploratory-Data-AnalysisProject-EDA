# ***EDA***

```python
df.columns
```

```python
sns.set(rc= {'figure.figsize' : (8,5)})
```

**GENDER**

```python
ax = sns.countplot(x = 'Gender', data= df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by = 'Amount')
sns.barplot(x='Gender', y='Amount',data = sales_gen)
```

***From above graphs we understand that females are shopping more than males and female have more purchasing power.***

**AGE**

```python
ax = sns.countplot(x = 'Age Group', data= df, hue ='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by = 'Amount')
sns.barplot(x='Age Group', y='Amount',data = sales_age)
```

***Women in age group 26-35 have spent the most.***

**STATE**

```python
sns.set(rc= {'figure.figsize' : (15,5)})

ax = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending= False).head(10)
sns.barplot(x='State', y='Amount',data = ax)
```

```python
sns.set(rc= {'figure.figsize' : (15,5)})

ax = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.barplot(x='State', y='Orders',data = ax)
```

***People in UP, Maharashta and Karnata bought more and have more purchasing power.***

**MARITAL STATUS**

```python
sns.set(rc= {'figure.figsize' : (8,5)})
aa = sns.countplot(x = 'Marital_Status', data= df, hue= 'Gender')

for bars in aa.containers:
  aa.bar_label(bars)
```

```python
sales_ms = df.groupby(['Marital_Status','Gender'], as_index= False)['Amount'].sum().sort_values(by = 'Amount', ascending= False)
sns.barplot(data = sales_ms, x= 'Marital_Status', y= 'Amount', hue = 'Gender')
```

***Married women are most likely to do Diwali shopping.***

**OCCUPATION**

```python
aa = sns.countplot(x = 'Occupation', data= df)

for bars in aa.containers:
  aa.bar_label(bars)
```

```python
sns.set(rc= {'figure.figsize' : (20,5)})
sales_occ = df.groupby(['Occupation'], as_index= False)['Amount'].sum().sort_values(by = 'Amount', ascending= False)
sns.barplot(data = sales_occ, x= 'Occupation', y= 'Amount')
```

***From above graphs we understand most buyes are from IT, Healthcare and Aviation, and their buying power is also in the same order.***

**PRODUCT CATEGORY**

```python
aa = sns.countplot(x = 'Product_Category', data= df)

for bars in aa.containers:
  aa.bar_label(bars)
```

```python
sns.set(rc= {'figure.figsize' : (15,5)})
sales_occ = df.groupby(['Product_Category'], as_index= False)['Amount'].sum().sort_values(by = 'Amount', ascending= False).head(10)
sns.barplot(data = sales_occ, x= 'Product_Category', y= 'Amount')
```

***People spend most on food, clothing and electronics in this order.***

**PRODUCT ID**

```python
sns.set(rc= {'figure.figsize' : (15,5)})
sales_pid = df.groupby(['Product_ID'], as_index= False)['Orders'].sum().sort_values(by = 'Orders', ascending= False).head(10)
sns.barplot(data = sales_pid, x= 'Product_ID', y= 'Orders')
```

# **CONCLUSION**

### ***Married women age group 26-35, from UP, Maharashtra and Karnata who working in IT, Healthcare and Aviation are more like to purchase products, specially in Food, Clothing and Tech products.***
